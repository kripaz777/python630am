```python
# function with argument and no return type

def info(name,age,add):
    x = f"Hello I am {name}. I am from {add}. I am {age}"
    print(x)


name = input("Enter name = ")
age = int(input("Enter age = "))
add = input("Enter add = ")

info(name,age,add)
```

```python
# function with argument and no return type

def info(x,y,z):
    a = f"Hello I am {x}. I am from {y}. I am {z}"
    print(a)


name = input("Enter name = ")
age = int(input("Enter age = "))
add = input("Enter add = ")

info(name,age,add)
```

```python
def cal():
    l = int(input("Enter l = "))
    b = int(input("Enter b = "))
    h = int(input("Enter h = "))
    
    a = l*b
    v = a*h
    return a,v

area,volume = cal()
print("The area = ",area)
print("The volume = ",volume)
```

```python
#function with argument and return type
def cal(l,b,h): 
    a = l*b
    v = a*h
    return a,v

l = int(input("Enter l = "))
b = int(input("Enter b = "))
h = int(input("Enter h = "))

area,volume = cal(l,b,h)
print("The area = ",area)
print("The volume = ",volume)
```
