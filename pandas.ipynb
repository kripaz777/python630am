```python
import csv
info = []
csv_columns = ['name', 'age', 'add']
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter add = ")
    info.append({'name':name,'age':age,'add':add})
    
try:
    with open("data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in info:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
